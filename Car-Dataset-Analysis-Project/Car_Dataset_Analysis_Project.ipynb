## 🚗 Dataset Description

* Make: The manufacturer of the car (e.g., Toyota, Ford, BMW).
* Model: The specific model name of the car.
* Year: The production year of the car.
* Engine Fuel Type: Type of fuel the car uses (e.g., gas, diesel, electric, hybrid).
* Engine HP: The horsepower of the car's engine.
* Engine Cylinders: The number of cylinders in the car's engine.
* Transmission Type: The type of transmission in the car (e.g., automatic, manual).
* Driven_Wheels: The type of drivetrain (e.g., front-wheel drive, rear-wheel drive, all-wheel drive).
* Number of Doors: Number of doors the car has (e.g., 2, 4).
* Market Category: The market segment the car belongs to (e.g., SUV, sedan, luxury).
* Vehicle Size: The size category of the car (e.g., compact, mid-size, large).
* Vehicle Style: The style of the vehicle (e.g., coupe, convertible, hatchback).
* highway MPG: The estimated miles per gallon on the highway.
* city mpg: The estimated miles per gallon in the city.
* Popularity: A numerical value indicating how popular the car is in the market.
* MSRP: The Manufacturer's Suggested Retail Price of the car.

## 🧹 Data Cleaning Process
Data cleaning is the next step and helps to correct errors, inconsistencies and inaccuracies to ensure that the data we analyze is accurate, complete and reliable. We do this using the following steps:

* Imputation - replace the missing value with either the mean or mode of that column
* Deleting the null values (only if the portion of missing data is small)
* Check data types and correct where necessary (in our case all are already correct)
* Removing duplicates
* if any data types incorrect then to do data type conversions (not for the car_df as all correct)
* Cleaning any strings where necessary (check capitalization, remove white space) - as per instructions change Vehicle Style and Market Category entries to lowercase
* Create new columns
* Filter data as per instruction to only include from 1995 onwards


#### Missing data -  
- Replace with median - 69 in HP
- Replace with 'missing data' - 3742 in Market Category
- Delete rows - 3 in Engine Fuel Type, 6 in number of doors and 30 in Cylinders

```python
# replace missing values with median in Engine HP
car_df['Engine HP'] = car_df['Engine HP'].fillna(car_df['Engine HP'].median())

# remove column
car_df['Market Category'] = car_df['Market Category'].fillna('Missing Data')

# Remove rows with missing vals - Engine Fuel Type, Engine Cylinders and Number of Doors
car_df = car_df.dropna(axis=0)

# Check for null values in all columns
car_df.isnull().sum()
```

#### Remove duplicates

```python
# Shape of sataframe is 11875 rows and 16 columns to start
car_df = car_df.drop_duplicates()

# Reprint shape to chack if anything removed - now rows down to 11160, therefore 715 duplicate rows removed
car_df.shape
```

#### String operations

```python
# Change the text in Vehicle Style and Market Category to lower case

car_df['Vehicle Style'] = car_df['Vehicle Style'].str.lower()
car_df['Market Category'] = car_df['Market Category'].str.lower()
car_df
```

#### Create new Column
As per the instructions, create a new columns called
- 'Total MPG' that is the average of 'city mpg' and 'highway MPG.'
- 'Price per HP' that is 'MSRP' divided by 'Engine HP

```python
# Create new col and add 2 MPG cols together and divide by 2 to get the avg
car_df['Total MPG'] = (car_df['city mpg'] + car_df['highway MPG'])/2

# Create new col and divide the 2 cols by each other, changed data type to int to match the MSRP as it has no decimal places
car_df['Price per HP'] = car_df['MSRP'] / car_df['Engine HP']
car_df['Price per HP'] = car_df['Price per HP'].astype(int)

car_df
```

#### Filter the data

As per the project guidelines only data from 1995 onwards is needed, so created a new data set called car_df_filtered so that I dont override the original database
* Did this last as that way all the data cleaning and manipulation happened on origonal database incase we ever want to filter different years at a later stage

```python
car_df_filtered = car_df[car_df['Year'] > 1994]
car_df_filtered
```

```python
# Check only correct years in the new df
car_years = car_df_filtered['Year'].unique()
print(np.sort(car_years))
```

2. Group Analysis:
Group the data by the following columns and calculate the average MSRP and Popularity for each group:
- Driven_Wheels
- Vehicle Size
- Engine Cylinders

```python
# Define a function to avoid code repetition

def group_and_mean(group_col, mean1_col, mean2_col):
    return car_df_filtered.groupby(group_col)[[mean1_col, mean2_col]].mean().round(2)
```

```python
# Group Drive Wheels
driven_wheels_agg = group_and_mean('Driven_Wheels', 'MSRP', 'Popularity')
driven_wheels_agg
```

```python
# Group Vehicle Size
vehicle_size_agg = group_and_mean('Vehicle Size', 'MSRP', 'Popularity')
vehicle_size_agg
```

```python
# Group Vehicle Size
engine_cyl_agg = group_and_mean('Engine Cylinders', 'MSRP', 'Popularity')
engine_cyl_agg
```

3. Visualizations:
Generate the following visualizations:
- A histogram that shows a distribution for the city mpg column.
- A bar chart showing the average MSRP for each category in Vehicle Size.
- A scatter plot showing the relationship between Engine HP and MSRP.
- A boxplot showing the distribution of MSRP for each category in Driven_Wheels.
- A line plot showing trends in average city mpg and highway MPG for different Transmission Type.

```python
# (Fig 1) Create a histogram for the city mpg column - right skewed

mpg_column = car_df_filtered['city mpg']

plt.figure(figsize=(10,6))
plt.hist(mpg_column, bins=30, alpha=0.7, edgecolor='darkgreen', color='lightgreen')
plt.xlabel('Bins')
plt.ylabel('Values')
plt.title('Histogram of City MPG')
plt.show()
```

```python
# (Fig 2) bar chart showing the average MSRP for each category in Vehicle Size
plt.figure(figsize=(10,6))
sns.barplot(x='Vehicle Size', y='MSRP', data=car_df_filtered, estimator='mean', errorbar=None, color='lightgreen')

plt.xlabel('Vehicle Size')
plt.ylabel('Avg MSRP')
plt.title('Avg MSRP for each category in Vehicle Size')
plt.show()
```

```python
# (Fig 3) scatter plot showing the relationship between Engine HP and MSRP
plt.figure(figsize=(10,6))
sns.scatterplot(data=car_df_filtered, x='Engine HP', y='MSRP', color='green')
plt.xlabel('Engine HP')
plt.ylabel('MSRP')
plt.title('Relationship between Engine HP and MSRP')
plt.show()
```

```python
# (Fig 4) A boxplot showing the distribution of MSRP for each category in Driven_Wheels.
plt.figure(figsize=(10,10))
sns.boxplot(data=car_df_filtered, x='Driven_Wheels', y='MSRP')
plt.xlabel('Driven Wheels')
plt.ylabel('MSRP')
plt.title('Distribution of MSRP for each category in Driven_Wheels')
plt.show()
```

4. Correlation Analysis:
- Investigate the correlation between the following variables:
    * `Engine HP`, `MSRP`, `Popularity`, `city mpg`, and `highway MPG`

```python
# (Fig 5) create a heatmap to show correlation between all the numeric variables
numeric_data = car_df_filtered.loc[:, ['Engine HP', 'MSRP', 'Popularity', 'city mpg', 'highway MPG']]
corr_matrix = numeric_data.corr(method='pearson')

plt.figure(figsize=(10,6))
sns.heatmap(data=corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
```

```python
# (Fig 6) Scatter plot to check correlation between highway MPG and Engine HP
plt.figure(figsize=(10,6))
sns.scatterplot(data=car_df_filtered, x='Engine HP', y='highway MPG', hue='Vehicle Size')
plt.xlabel('Engine HP')
plt.ylabel('Highway MPG')
plt.title('Relationship between Engine HP and Highway MPG split by Vehicle Size')
plt.show()
```

```python
# (Fig 7) Scatter plot to check correlation between highway MPG and city MPG
plt.figure(figsize=(10,6))
sns.scatterplot(data=car_df_filtered, x='city mpg', y='highway MPG', hue='Transmission Type')
plt.xlabel('City MPG')
plt.ylabel('Highway MPG')
plt.title('Relationship between City MPG and Highway MPG split by Transmission Type')
plt.show()
```
